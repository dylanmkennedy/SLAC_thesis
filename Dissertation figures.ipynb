```python
def f(xs, a=3, b=10, sigma=0.15, noise=True):
    ys = np.exp((-a*(xs-2)**2*(xs+2)**2 + b*xs) / 60)
    if noise:
        ys += np.array([sigma*np.random.randn() for i in range(len(ys))])
    return ys

def f2(xs, noise=True):
    ys = np.exp(-xs**2/2)
    if noise:
        ys += 0.05*np.array([sigma*np.random.randn() for i in range(len(ys))])
    return ys

def optimizer(obj_func, initial_theta, bounds):
    res = l_bfgs_b(obj_func, initial_theta, bounds=bounds)
    return res[0], res[1]


def EI(mu, sigma, ybest):
    Z = (mu-ybest)/sigma
    return (mu - ybest)*norm.cdf(Z)+sigma*norm.pdf(Z)

def PI(mu, sigma, ybest):
    return (1 - norm.cdf(ybest, mu, sigma))

def center_axes(ax1):
    # Move left y-axis and bottim x-axis to centre, passing through (0,0)
    ax1.spines['left'].set_position('center')
    ax1.spines['bottom'].set_position('center')

    # Eliminate upper and right axes
    ax1.spines['right'].set_color('none')
    ax1.spines['top'].set_color('none')

    # Show ticks in the left and lower axes only
    ax1.xaxis.set_ticks_position('bottom')
    ax1.yaxis.set_ticks_position('left')
    ax1.set_xticklabels([])
    ax1.set_yticklabels([])

def center_axes2(ax1):
    # Move left y-axis and bottim x-axis to centre, passing through (0,0)
    ax1.spines['left'].set_position('center')

    # Eliminate upper and right axes
    ax1.spines['right'].set_color('none')
    ax1.spines['top'].set_color('none')

    # Show ticks in the left and lower axes only
    ax1.xaxis.set_ticks_position('bottom')
    ax1.yaxis.set_ticks_position('left')
    ax1.set_xticklabels([])
    ax1.set_yticklabels([])

#needed to define a wrapper function for l_bfgs_b algorithm (gets used to fit the GP) to be able to adjust the maxiter.
import functools
from scipy.optimize import fmin_l_bfgs_b
from scipy.stats import norm

l_bfgs_b = functools.partial(fmin_l_bfgs_b, maxiter=100000)



from sklearn.gaussian_process import GaussianProcessRegressor as gpr
from sklearn.gaussian_process.kernels import RBF, WhiteKernel, ConstantKernel


```

```python
#amplitude hyperparam

w=8
numpoints = 1
xs_meas = np.array([(w*np.random.random()-w/2) for i in range(numpoints)])

ys_meas = f(xs_meas)

xs = np.linspace(-w/2., w/2., 1000)
ys = f(xs, noise=False)


X_train = np.array([[x] for x in xs_meas])
#rescale the input data before fitting the gp
from sklearn import preprocessing
scaler = preprocessing.StandardScaler().fit(X_train)
X_scaled = scaler.transform(X_train)





# Create two subplots and unpack the output array immediately
fig, axes = plt.subplots(1, 2, sharey=True)

for i in [0,1]:
    ax = axes[i]
    center_axes(ax)
    ax.set_xlabel(r'$x$', loc = 'right')
    ax.set_ylabel(r'$y$', loc = 'top', rotation = 'horizontal')
    
    amp = [0.7,2][i]
    length=0.7
    noise = 0.15**2
    # noise = 10**2
    lbounds = 'fixed'
    # lbounds = (0.01,20)
    nbounds = 'fixed'
    # nbounds = (1.e-5, 1.e5)
    # amp = 1
    amp_bounds = "fixed"
    # amp_bounds = (1.e-5, 1.e5)
    kernel = ConstantKernel(amp**2, constant_value_bounds=amp_bounds)*RBF(length, length_scale_bounds=lbounds) + WhiteKernel(noise_level=noise, noise_level_bounds = nbounds)
    gp = gpr(kernel=kernel, optimizer = optimizer, n_restarts_optimizer=100)
    gp.fit(X_scaled, ys_meas)


    xs = np.linspace(-w/2., w/2., 1000)
    # xs = np.linspace(-w, w, 1000)

    ys = f(xs, noise=False)
    ax.plot(xs,ys, c='k', label = 'Target Function')

    X_pred = [[x] for x in xs]
    X_pred = scaler.transform(X_pred)
    ys_pred, sigmas_pred = gp.predict(X_pred, return_std = True)
    ax.plot(xs, ys_pred, "b--", label="GP MLE")
    ax.fill(
        np.concatenate([xs, xs[::-1]]),
        np.concatenate([ys_pred - 1.9600 * sigmas_pred, (ys_pred + 1.9600 * sigmas_pred)[::-1]]),
        alpha=0.3,
        fc="b",
        ec="None",
        label="GP 95% CI",)

    ax.scatter(xs_meas, ys_meas, label = 'Observations')
    ax.legend(loc = 'lower left')
plt.ylim(-4,4)

```

```python
#iterations of Bayes opt

w=6
numpoints = 1
xs_meas = np.array([(w*np.random.random()-w/2) for i in range(numpoints)])

ys_meas = f(xs_meas)








niter = 3

fig, axes = plt.subplots(2, niter, sharey='row')
fig.set_figheight(4)
fig.set_figwidth(8)

max_acq = []

iters = range(0,niter)
for i in iters:
    
    X_train = np.array([[x] for x in xs_meas])
    #rescale the input data before fitting the gp
    from sklearn import preprocessing
    scaler = preprocessing.StandardScaler().fit(X_train)
    X_scaled = scaler.transform(X_train)
    
    
    lbounds = 'fixed'
#     lbounds = (0.01,20)
    nbounds = 'fixed'
#     nbounds = (1.e-5, 1.e5)
    amp_bounds = "fixed"
#     amp_bounds = (1.e-5, 1.e5)

    ax = axes[1,i]
    center_axes(ax)
    ax.set_xlabel(r'$x$', loc = 'right')
    ax.set_ylabel(r'$y$', loc = 'top', rotation = 'horizontal')
    ax.set_ylim(-4,4)
    
    length=1
    noise = 0.15**2
    amp = 0.7
    kernel = ConstantKernel(amp**2, constant_value_bounds=amp_bounds)*RBF(length, length_scale_bounds=lbounds) + WhiteKernel(noise_level=noise, noise_level_bounds = nbounds)
    gp = gpr(kernel=kernel, optimizer = optimizer, n_restarts_optimizer=100)
    gp.fit(X_scaled, ys_meas)


#     xs_pred = np.linspace(-w/2., w/2., 1000)
    xs_pred = np.linspace(-w, w, 1000)

    ys = f(xs_pred, noise=False)
    ax.plot(xs_pred,ys, c='k', label='Target Function')

    X_pred = [[x] for x in xs_pred]
    X_pred = scaler.transform(X_pred)
    ys_pred, sigmas_pred = gp.predict(X_pred, return_std = True)
    ax.plot(xs_pred, ys_pred, "b--", label="GP MLE")
    ax.fill(
        np.concatenate([xs_pred, xs_pred[::-1]]),
        np.concatenate([ys_pred - 1.9600 * sigmas_pred, (ys_pred + 1.9600 * sigmas_pred)[::-1]]),
        alpha=0.3,
        fc="b",
        ec="None",
        label="GP 95% CI",)

    ax.scatter(xs_meas, ys_meas, label='Observations')

    
    ybest = np.max(ys_meas)
    acq = EI(ys_pred, sigmas_pred, ybest)
    max_acq += [np.max(acq)]
    x_acq = xs_pred[np.argmax(acq)]
    y_acq = f(np.array([x_acq]))
#     ax.scatter([x_acq],y_acq,c='g', marker='x')
    ax.scatter([x_acq],[0],c='g', marker='x')
    
    ax = axes[0,i]
    center_axes2(ax)
    ax.set_xlabel(r'$x$', loc = 'right')
    ax.set_ylabel(r'$E.I.$', loc = 'top', rotation = 'horizontal')
    ax.plot(xs_pred, acq, 'g-', label='EI')
    ax.scatter(x_acq,np.max(acq),c='g', marker='*')
    
    xs_meas = np.array(list(xs_meas) + [x_acq])
    ys_meas = np.array(list(ys_meas) + list(y_acq))
    ax.set_ylim(bottom=0)
for i in iters:
    ax = axes[0,i]
    ax.text(-6,0.015, 'iter# ' + str(i+numpoints+1))

    plt.tight_layout()
```
